# import packages for plotting graphs and manipulating data:¶

```python
import numpy as np
import matplotlib.pyplot as plt
```

# define my function1: f(x1)= 3x^2 - 2x + 1

```python
def myFunction(x):
    f = 3 * x ** 2 - 2 * x + 1
    return f
```

# define the derivative of my function1: f ′ (x)=6x - 2

```python
def myDerivativeFunction(x):
    Df = 6 * x - 2
    #Df=Derivative(function, f)
    return Df
```

# define the domain of the function: x=[−20:0.1:20] 

```python
x = np.arange(-10, 10, 0.1
```

# pick 3 points in the domain

```python
a1= 0
a2= -4
a3= 3
#each of a means the x-value of each point
```

# Taylor approximation

```python
def taylorFunction(a):
    Tf = myFunction(a) + myDerivativeFunction(a) * (x-a)
    return Tf
#use the function what I learned in the class: taylor funtion TF= f(x1)+f'(x1)(x-x1)
```

# compute the graph

```python
f = myFunction(x)
Df = myDerivativeFunction(x)
Tf1 = taylorFunction(a1)
Tf2 = taylorFunction(a2)
Tf3 = taylorFunction(a3)
```

# plot the graphs for the function and its derivative

```python
plt.figure(1)
plt.plot(x, f, 'b', label="function")
plt.plot(x, Df, 'r', label="derivative")
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```

# plot the graphs of Taylor approximation at the given points

```python
plt.figure(2)
plt.plot(x, f, 'b', label="function")
plt.plot(x, Tf1, 'm', label="Taylor approximation at a1")
plt.plot(x, Tf2, 'g', label="Taylor approximation at a2")
plt.plot(x, Tf3, 'y', label="Taylor approximation at a3")
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```
